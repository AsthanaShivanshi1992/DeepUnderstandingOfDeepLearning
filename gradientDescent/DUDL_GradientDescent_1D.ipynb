# Gradient descent in 1D

```python
#function 
def fx(x):
    return 3*x**2- 3*x+4

#function for the derivative of the above function
def deriv(x):
    return 6*x-3 #Manually, hard coding without any referencing
```

```python
#Plotting the fucntion and the derivative
x=np.linspace(-2,2,2001)# x takes values between -2 and 2 with 2001 steps (learnign rate analogy, to be discussed later)
#plotting
plt.plot(x,fx(x),x,deriv(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','fx'])
plt.show()
```

```python
# random starting point
localmin = np.random.choice(x,1) #Just taking 1 random value from x
print(localmin)

# learning parameters
learning_rate = .01 #steps taken along the graph, has to be small so we can catch what we want
training_epochs = 100 #number of iterations to repeat the entire thing in the model so it learns

# run through training
for i in range(training_epochs):
  grad = deriv(localmin)
  localmin = localmin - learning_rate*grad #Gradient descent algorithm= our starting point minus product of derivof the starting point and learning rate

localmin
```

```python
#Plotting the results
plt.plot(x,fx(x),x,deriv(x))
plt.plot(localmin,deriv(localmin),'ro') #'ro': This string specifies the style of the plotted points:r: Stands for "red," indicating the color of the points.o: Represents the marker style, in this case, circles.
plt.plot(localmin,fx(localmin),'ro')

plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x) min'])
plt.title('Empirical local minimum: %s'%localmin[0])
plt.show()
```

# Store the model parameters and outputs on each iteration

```python
#random starting point as before
localmin=np.random.choice(x,1)
#Learning parameters
learning_rate=.01
training_epochs=100

#run through training and store all the results for each step
#initialising an empty array fro storing the results
modelparams=np.zeros((training_epochs,2)) #Has rowsx columns, number of rows=no. of training epochs, number of columns , 2
for i in range(training_epochs):
    grad=deriv(localmin)
    localmin=localmin-learning_rate*grad
    modelparams[i,0]=localmin #first column stores localmin for each training epoch
    modelparams[i,1]=grad #Second column stores gradient(deriv) of the localmin

```

```python
#going to plot both the columns as subplots now
```

```python
#Two subplots: one showing the local min and how it changes and another showing the derivative
# plot the gradient over iterations

fig,ax = plt.subplots(1,2,figsize=(12,4))
#running a loop to plot information from both the columns
for i in range(2): #Takes columns 1 and 2
  ax[i].plot(modelparams[:,i],'o-')
  ax[i].set_xlabel('Iteration: Training epochs')
  ax[i].set_title(f'Final estimated minimum: {localmin[0]:.5f}') #Printing the local min accurate to five decimal places

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```
